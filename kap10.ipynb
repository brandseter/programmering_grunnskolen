# Eksempel: Beregning av $\pi$ 

- $\pi$ er definert som **forholdet** mellom *omkretsen* til en sirkel og *diameteren*: $\pi =\dfrac{O}{d}=\dfrac{O}{2r}$
- Vi tegner en sirkel med diameter 1. Da er$\pi=O$. 
- Da vet vi at verdien til $pi$ må ligge et sted mellom omkretsen til den *innskrevede* konvekse mangekanten og den *omskrevne* mangekanten

Arkimedes algoritme for å tilnærme pi

Men det er vanskelig å beregne omkretsen til disse mangekantene. 

# 10.1.3 Beregning av $\pi$ med tilfeldige tall

$\color{red}{A_s = \pi r^2\\}$
$\color{blue}{A_k = (2r)^2}$

$\dfrac{\color{red}{A_s}}{\color{blue}{A_k}}=\dfrac{\color{red}{\pi r^2}}{\color{blue}{(2r)^2}}=\dfrac{\pi}{4}$
